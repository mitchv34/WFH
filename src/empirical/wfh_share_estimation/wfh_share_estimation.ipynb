```python
import numpy as np
from scipy.special import logit, expit
from sklearn.datasets import make_regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

# Scale the target to be between 0.001 and 0.999
X_train = pipeline3.data.X_train
X_test = pipeline3.data.X_test
y_train = pipeline3.data.y_train
y_test = pipeline3.data.y_test
iz_train = pipeline3.data.iz_train
iz_test = pipeline3.data.iz_test

# Train the regressor on non-zero data only.
X_nonzero = pipeline3.data.X_train[pipeline3.data.iz_train != 1]
y_nonzero = pipeline3.data.y_train[pipeline3.data.iz_train != 1]

# Apply logit transformation to the target variable
# This transforms our (0,1) bounded data to unbounded data that can take any real value
y_train_transformed = logit(y_nonzero)

# Add constant term for the intercept
X_train_sm = sm.add_constant(X_nonzero)

# Fit linear regression on transformed data
model = sm.OLS(y_train_transformed, X_train_sm)
results = model.fit()
```

```python
# Stage 1: Use the classifier to predict zeros.
zero_probs = pipeline3.calibrated_classifier.predict_proba(X_test)[:, 1]
predicted_zero = (zero_probs > pipeline3.zero_threshold).astype(int)

X_non_zero = X_test.loc[predicted_zero != 1]
X_test_sm = sm.add_constant(X_non_zero)

# Make predictions
# First get the raw predictions
raw_predictions = results.predict(X_test_sm)
# Then transform back to (0,1) scale using expit (inverse logit)
final_pred = np.zeros(len(pipeline3.data.X_test))
final_pred[predicted_zero == 1] = 0
final_pred[predicted_zero != 1] = expit(raw_predictions)
```

```python
# Calculate performance metrics
from sklearn.metrics import mean_squared_error, r2_score

f1_score = pipeline3.scores['f1']
mse = mean_squared_error(y_test, final_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, final_pred)

print(f"F1 Score: {f1_score:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"R-squared Score: {r2:.4f}")

results_df = y_test.to_frame().assign(Predicted=final_pred)
results_df["Actual_Zero"] = results_df["ESTIMATE_WFH_ABLE"].apply(lambda x: 1 if x > 0 else 0)
results_df["Predicted_Zero"] = results_df["Predicted"].apply(lambda x: 1 if x > 0 else 0)
results_df['Correct_Class'] = results_df["Actual_Zero"] == results_df["Predicted_Zero"]
        
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(x="ESTIMATE_WFH_ABLE", y="Predicted", data=results_df, ax=ax,
                s=200, hue='Correct_Class')
ax.set_xlabel("Actual WFH Estimate")
ax.set_ylabel("Predicted WFH Estimate")
ax.set_title("Actual vs. Predicted WFH Estimate")
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls="--", c="gray")
sns.despine()
```

```python
import numpy as np
from sklearn.base import BaseEstimator, RegressorMixin
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from statsmodels.othermod.betareg import BetaModel

class BetaRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, add_intercept=True, link=sm.families.links.logit()):
        self.add_intercept = add_intercept  # Control intercept addition
        self.link = link
        self.model = None
        self.result = None
        
    def fit(self, X, y):
        # Ensure `y` is in (0, 1)
        y = np.clip(y, 1e-6, 1 - 1e-6)
        
        # Add intercept column if requested
        if self.add_intercept:
            X = sm.add_constant(X)
            
        # Create and fit Beta regression model
        self.model = BetaModel(
            endog=y,
            exog=X,
        )
        self.result = self.model.fit()
        return self
    
    def predict(self, X):
        if self.add_intercept and not np.all(X[:, 0] == 1):
            X = sm.add_constant(X)
        return self.result.predict(X)

    def get_params(self, deep=True):
        return {"add_intercept": self.add_intercept, "link": self.link}

    def set_params(self, **parameters):
        for parameter, value in parameters.items():
            setattr(self, parameter, value)
        return self

# --------------------------------------------------
# Example Usage
# --------------------------------------------------
# Generate synthetic data (bounded between 0 and 1)
X, y = make_regression(n_samples=1000, n_features=5, noise=0.1)
y = MinMaxScaler((0.001, 0.999)).fit_transform(y.reshape(-1, 1)).flatten()

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create pipeline with intercept handling
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),  # Scale features
    ('beta', BetaRegressor(add_intercept=True))  # Beta regressor adds intercept
])

# Train and predict
pipeline.fit(X_train, y_train)
predictions = pipeline.predict(X_test)
```

```python
# Load Mongey Pilossoph and Weinberg (2020) classification
low_wfh_pp = pd.read_csv("/project/high_tech_ind/WFH/WFH/data/aux_and_croswalks/low_wfh_pp.csv", usecols=["OCC_CODE", "high_wfh", "low_wfh_binary"])

onet_soc_xwalk = pd.read_excel("https://www.onetcenter.org/taxonomy/2019/soc/2019_to_SOC_Crosswalk.xlsx?fmt=xlsx", skiprows=3)
onet_soc_xwalk.rename(
    columns= lambda x: x.replace(' ','_').replace('-','_').replace('*','').replace("2019_","").replace("2018_","").upper(), 
    inplace=True
    )
onet_soc_xwalk.rename(columns={'SOC_CODE':'OCC_CODE', 'SOC_TITLE':'OCC_TITLE'}, inplace=True)

low_wfh_pp = low_wfh_pp.merge(onet_soc_xwalk, on="OCC_CODE", how="left")[["ONET_SOC_CODE", "ONET_SOC_TITLE", "low_wfh_binary", "high_wfh"]]
low_wfh_pp = low_wfh_pp.merge(data_store.labeled_data[["ESTIMATE_WFH_ABLE"]].reset_index(), on="ONET_SOC_CODE", how="outer", indicator=True)
low_wfh_pp_labeled = low_wfh_pp[low_wfh_pp["_merge"]=="both"].set_index("ONET_SOC_CODE")
```

```python
# Compute an F1 score for the classification using low_wfh_binary 
from sklearn.metrics import f1_score
# Select only an equivalent to our test set
low_wfh_pp_labeled_test = low_wfh_pp_labeled.loc[data_store.X_test.index]
f1_score_mpw = f1_score(low_wfh_pp_labeled_test["low_wfh_binary"], 
                        low_wfh_pp_labeled_test["ESTIMATE_WFH_ABLE"] == 0.0)
f1_pipeline1 = pipeline1.scores["f1"]
f1_pipeline2 = pipeline2.scores["f1"]
f1_pipeline3 = pipeline3.scores["f1"]
print(f"F1 score:")
print(f"\tMongey, Pilossoph, and Weinberg (2020) classification: {f1_score_mpw}")
print(f"\t(pipeline1): {f1_pipeline1}")
print(f"\t(pipeline2): {f1_pipeline2}")
print(f"\t(pipeline3): {f1_pipeline3}")
```
